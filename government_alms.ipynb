## Análise e Visualização dos Dados:
---

>Após obtermos os dados estruturados, utilizamos as bibliotecas **Numpy**, **Pandas** e **Matplotlib** para a exploração dos dados:

```python
from util.plot import Plot

import pandas as pd
data = pd.read_json('./api.json')
aux = pd.read_json('./parameters/estados.json')

plt = Plot(data)
```

## Crecimento Médio Anual por Estado:
---
> O gŕafico abaixo, faz um demonstrativo dos gastos em certo estado dentro do limite de tempo escolhido.

```python

plt.plot_timeline('BAHIA', 2004, 2007)
```

## Visualização da Frequência do Aumento dos Gastos Anual:
---
>O gráfico abaixo demonstra o gasto de certo estado em determinado ano.

```python

plt.plot_state_year('SERGIPE', 2007)
```

## Gráfico Comparativo dos Gastos por Período:
---
>Faz o comparativo de gastos entre dois estados em um determinado período.

```python

plt.verify_mean('BAHIA', 'RORAIMA', 2004, 2007)
```

## Visualização dos Gastos em Percentual por Estado:
---
>O Próximo gŕafico, demonstra a porcentagem dos gastos no estado e período de tempo escolhido. 

```python
plt.plot_percent_year('RORAIMA', 2004, 2014)
```
